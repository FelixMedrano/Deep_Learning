```
 from keras.datasets import mnist
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# Set variables to the approriate places
(x_train,y_train),(x_test,y_test)=mnist.load_data()

#Part a
# print("The amount of images for x_train is:", x_train.shape[0])
# print("The amount of images for x_test is:", x_test.shape[0])
filtered=(y_train==0) | (y_train==1)
x_train=x_train[filtered]
y_train=y_train[filtered]

filtered=(y_test==0) | (y_test==1)
x_test=x_test[filtered]
y_test=y_test[filtered]

# print("The amount of images for x_train is:", x_train.shape[0])
# print("The amount of images for x_test is:", x_test.shape[0])





#List to store the x values of graph, y values(avg pixel values), and whether it was a 0 or 1
xVal=[]
yVal=[]
Binary=[]
# using j to get random int
j=0
#creating validation set
x_valid=x_train[0:x_train.shape[0],:,:]

#loop to get random 500 images
for i in range(500):
  
  #fill set with 1-500 x values and get random index 
  xVal.append(i)
  j=np.random.randint(0,x_train.shape[0])
 
  #get pixel value, find avg and add to list
  x=(x_valid[j,13:16,13:16])
  avg=np.mean(x)
  yVal.append(avg)
  
  #Record whether it was a 1 or a 0
  Binary.append(y_train[j])
  
  


#Variables for accuracy calculation
OneCount=0
ZeroCount=0
#This loop plots based on if they were 0's or 1's
for c in range(500):
  if Binary[c]==1:
    plt.scatter(xVal[c], yVal[c], marker='v', color='red')
    #Set the threshold
    if yVal[c]>=100:
      OneCount+=1
    # print("its 1")
  else:
    plt.scatter(xVal[c], yVal[c], marker='o', color='blue')
    #set the threshold
    if yVal[c]<100:
      ZeroCount+=1
    # print("its 0")


#plot the graph
plt.title('Handwritten digits')
plt.xlabel("Image Sample")
plt.ylabel("Average pixel value")
Ones= mpatches.Patch(color='red', label='Ones')
Zeroes = mpatches.Patch(color='blue', label='Zeroes')
plt.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1, loc='lower right',handles=[Ones, Zeroes]);
plt.show()

#accuracy calculation
Acc=((OneCount+ZeroCount)/500)*100
print("If we take the threshhold to be a y value of about 100, the Accuracy of this is: ", Acc, "%")

```
