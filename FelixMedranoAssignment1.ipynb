```
import math
import numpy
import sys
#input first
while True:
  while True:
    #check for appropriate input
    try:
      FirstNum= input("Please Enter First Number:  ")
      #check for stop code
      if FirstNum=="x"or FirstNum=="X":
          print("Termination Code detected")
          sys.exit(0)
      FirstNum=float(FirstNum)
    except ValueError:
        print("Provide an Integer value")
        continue
    break;

#input what the operation will be
  while True:
     try:
      Sign=input("Please enter a 1(Addition), 2(Subtraction), 3(Multiplication), 4(Division), 5(Power), 6 (Mod), 7(EXP), 8(ln), 9(Abs):  ")
      #check for stop code
      if Sign=="x"or Sign=="X":
          print("Termination Code detected")
          sys.exit(0)
      Sign=int(Sign)
      #check input is one of the options
      if Sign<1 or Sign>9:
        print("Error not within scope")
        continue
     except ValueError:
       print("Provide an Integer value from the Choices given")
       continue
     break;

#input second number
  while True:
    try:
      #check to see if operation requires second variable
     if Sign==7 or Sign==8 or Sign==9:
      break;
     SecondNum= input("Please Enter Second Number:  ")
     #check for stop code
     if SecondNum=="x"or SecondNum=="X":
          print("Termination Code detected")
          sys.exit(0)
     SecondNum=float(SecondNum)
    except ValueError:
        print("Provide an Integer value")
        continue
    break;
    #functions for calculations with 2 variables
  def sum(FirstNum,SecondNum):
   return float(FirstNum)+float(SecondNum)
  def sub(FirstNum,SecondNum):
   return float(FirstNum)-float(SecondNum)
  def mul(FirstNum,SecondNum):
   return float(FirstNum)*float(SecondNum)
  def div(FirstNum,SecondNum):
   return float(FirstNum)/float(SecondNum)
  def pow(FirstNum,SecondNum):
   return float(FirstNum)**float(SecondNum)
  def mod(FirstNum,SecondNum):
   return float(FirstNum)%float(SecondNum)
#print calculations
  if Sign==1:
   print("ADDITION: ")
   print(FirstNum, " + ", SecondNum, " = ", sum(FirstNum,SecondNum))
   continue
  if Sign ==2:
    print("SUBTRACTION: ")
    print(FirstNum, " - ", SecondNum, " = ", sub(FirstNum,SecondNum))
  if Sign ==3:
    print("MULTIPLICATION: ")
    print(FirstNum, " * ", SecondNum, " = ", mul(FirstNum,SecondNum))
  if Sign ==4:
   print("DIVISION: ")
   print(FirstNum, " / ", SecondNum, " = ", div(FirstNum,SecondNum))
  if Sign ==5:
   print("POWER: ")
   print(FirstNum, " ^ ", SecondNum, " = ", pow(FirstNum,SecondNum))
  if Sign ==6:
   print("MOD: ")
   print(FirstNum, " % ", SecondNum, " = ", mod(FirstNum,SecondNum))
  if Sign ==7:
    print("EXP: ")
    print("e^", FirstNum, " = ", "{:.2f}".format(math.exp(FirstNum)))
  if Sign ==8:
   print("NATURAL LOG: ")
   print("ln(", FirstNum,")", " = ", "{:.2f}".format(numpy.log(FirstNum)))
  if Sign ==9:
    print("ABSOLUTE VALUE: ")
    print("abs(", FirstNum,")", " = ", abs(FirstNum))

 
  
 
 
  
```
